**1. 데이터의 시각화**
 - 시각화를 통해 비즈니스의 인사이트를 파악해야함.
 - 그래프와 통계량은 대부분 정보(원본)을 요약하여 관점에 따라 결과가 달라짐.
   어떤 식으로든 요약하면 정보의 손실 발생

```python
import matplotlib.pyplot as plt
# 기본 형식 : plt.plot(x, y)
```

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2]} #딕셔너리로 묶어서 key값으로 줄 수도 있음
print(dict1)

plt.plot('v1', 'v2', '', data=dict1)
# x축, y축 값 지정 / 다른 옵션과 혼란 여지 없애기 위해 '' 넣음 생략 가능 / data= 는 딕셔너리, 데이터프레임 등을 지정
plt.xlabel('month') # x축 레이블
plt.ylabel('sales') # y축 레이블
plt.title('Monthly Sales') # 그래프 제목
plt.show()
```

**1) 라인 스타일 조정하기 (순서 상관 없음)**

- color
   - 'b' : blue
   - 'g' : green
   - 'r' : red
   - 'c' : cyan
   - 'm' : magenta
   - 'y' : yellow
   - 'k; : black
   - 'w' : white

---------------------------------

- linestyle
   - '-' : solid line style
   - '--' : dashed line
   - '-.' : dash-dot line
   - ':' dotted line
   
---------------------------------

- marker
   - '.' : point marker
   - ',' pixel 
   - 'o' : circle
   - 'v' : triangle_down
   - '^' : triangle_up
   - '<' : triangle_left
   - '>' : triangle_right
   - '1' : tri_down
   - '2' : tri up
   - '3' : tri_left
   - '4' : tri_right
   - 's' : square
   - 'p' : pentagon
   - '*' : star

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[3,5,2,1,4]}
plt.plot('v1', 'v2', 'b<-.', data=dict1)
plt.show()
```

**2) 여러 그래프 겹쳐서 그리기**
 - 그래프 두 개 그리고 한꺼번에 저장 후 출력
 
**3) 범례(Legend), 그리드 추가**

**4) 축 범위 조정하기 plt.xlim/ylim**
 - 그래프 범위 시작값과 끝값 지정
 

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
print(dict1)

plt.plot('v1', 'v2', 'go--', data=dict1, label='apple')
plt.plot('v1', 'v3', 'rs-', data=dict1, label='mango')

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly Sales')
plt.xlim(0,6)
plt.ylim(0,8)
plt.legend()
plt.grid()
plt.show()
```

**5) 그래프 크기 조정**



```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1', 'v3', 'rs-', data=dict1, label='mango')

plt.figure(figsize=(10,7))
plt.show()
```

**6) 수평선, 수직선 추가 plt.axhline(y값, color, linestyle)/axvline**

**7) 그래프에 텍스트 추가 plt.text(x좌표, y좌표, '추가할 텍스트')**

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1', 'v3', 'rs-', data=dict1, label='mango')
plt.axhline(5.3, color='grey', linestyle='--')
plt.axvline(2.4, color='grey', linestyle='--')

plt.text(2.5, 4.2, '2.4')
plt.show()
```

**8) 여러 그래프 나눠서 그리기 plt.subplot(rows, columns, index)**
 - 그래프간 간격 적절하 맞추기 : plt.tight_layout()

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.subplot(2,1,1)
plt.plot('v1', 'v2', 'go--', data=dict1)

plt.subplot(2,1,2)
plt.plot('v1', 'v3', 'rs-', data=dict1)

plt.tight_layout()
plt.show()
```
